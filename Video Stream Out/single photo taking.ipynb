```python
import cv2
import time
import os
from IPython.display import display
from PIL import Image

capture = cv2.VideoCapture(-1)
capture.set(3, 1280)
capture.set(4, 720)

frame_set = []
start_time = time.time()

while True:
    try:
        ret, frame = capture.read()
        rgb = cv2.flip(frame, flipCode = -1)
        cv2.imshow('frame', rgb)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        if time.time() - start_time >= 1: #<---- Check if 60x30 sec or 30 min passed
            img_name = "imgs/opencv_frame_{}.png".format(counter)
            cv2.imwrite(img_name, rgb)
            print("{} written!".format(counter))
            display(Image.open(img_name))
            start_time = time.time()
            counter += 1
            capture.release()
            break
    except:
        print("Camera 0 not found, switching to Camera 1")
        capture = cv2.VideoCapture(1)
        capture.set(3, 1280)
        capture.set(4, 720)         
```
